```python
from selenium.webdriver.common.by import By   # para buscar segun tag, id, class, etc...

import time

import warnings
warnings.filterwarnings('ignore')

from selenium.webdriver.chrome.options import Options


from webdriver_manager.chrome import ChromeDriverManager

from selenium import webdriver

import pandas as pd

import numpy as np
```

```python
PATH=ChromeDriverManager().install()   # instala el driver

driver=webdriver.Chrome(PATH)
```

```python
url = 'https://www.climbinghimalaya.com/total-open-peak-nepal.html'

driver.get(url)
```

```python
tabla=driver.find_element(By.TAG_NAME, 'tbody')

filas=tabla.find_elements(By.TAG_NAME, 'tr')


data=[]


for f in filas:
    
    elemen=f.find_elements(By.TAG_NAME, 'td')
    
    tmp=[]
    
    for e in elemen:
        
        tmp.append(e.text)
        
    data.append(tmp)

    
data
```

```python
data=[[e.text for e in f.find_elements(By.TAG_NAME, 'td')] 
      for f in tabla.find_elements(By.TAG_NAME, 'tr')]

data
```

```python
cabeceras=driver.find_element(By.TAG_NAME, 'thead')

cabeceras=[c.text for c in cabeceras.find_elements(By.TAG_NAME, 'th')]


cabeceras
```

```python
picos_loc=pd.DataFrame(data, columns=cabeceras)


picos_loc[:30]
```

```python
picos_loc.columns
```

```python
picos_loc.rename(columns={'Name  Height':'Name_height'}, inplace=True)
```

```python
picos_loc.to_csv('data/picos_loc.csv', index=False)
```
