```python
deaths.cause_of_death.unique() #Aquí rentaria reagrupar por causas y quitar la altitud como característica
```

```python
height_of_death = []

for i in deaths.cause_of_death:
    
    div = i.split()
    
    if len(div)==2:
        
        height_of_death.append(div[1])
        
    else:
        
        height_of_death.append(div[2])

deaths['height_of_death']= height_of_death

deaths.height_of_death
```

```python
deaths.height_of_death = deaths.height_of_death.str.replace('(', '')
deaths.height_of_death = deaths.height_of_death.str.replace(')', '')
deaths.height_of_death = deaths.height_of_death.str.replace('m', '')
deaths.height_of_death.head()
```

```python
deaths.rename(columns={'height_of_death': 'height_of_death_(m)'}, inplace=True)
```

```python
deaths['height_of_death_(m)']= deaths["height_of_death_(m)"].astype(dtype='int64', copy=True) 
# cambio el dato de de object a int porque son metros
```
