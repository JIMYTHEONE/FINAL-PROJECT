```python
height_of_death = []

for i in deaths.cause_of_death:
    
    div = i.split()
    
    if len(div)==2:
        
        height_of_death.append(div[1])
        
    else:
        
        height_of_death.append(div[2])

deaths['height_of_death']= height_of_death

deaths.height_of_death
```
